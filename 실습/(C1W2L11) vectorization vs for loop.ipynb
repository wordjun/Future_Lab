# vectorization vs for loop

vectorization은 기본적으로 코딩에서 명백한 for loop들을 제거하는 방식이다

```python
import numpy as np
```

```python
a = np.random.rand(1000000) #0과 1 사이의 값을 랜덤하게 갖는 100만 길이의 1-d array생성
b = np.random.rand(1000000)
print("Shape of a: " + str(a.shape))
print(b)
```

### for loop

```python
%%timeit
c=0
for i in range(len(a)):
    c += a[i]*b[i]
```

### vectorization

```python
%%timeit
np.dot(a,b)
```
